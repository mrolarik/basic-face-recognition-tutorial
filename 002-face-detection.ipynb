# Face Detection
Automatically find all the faces in an image  

useful link: http://face-recognition.readthedocs.io/en/latest/readme.html  

```python
import face_recognition

# Load the image file into a numpy array
image = face_recognition.load_image_file('face-image-test/unknown/u-06.jpg')
# Find all the faces in the image using the default HOG-based model.
face_locations = face_recognition.face_locations(image)

# face_locations is now an array listing the co-ordinates of each face!
print("Found {} face(s) in this photograph.".format(len(face_locations)))
```

```python
from skimage import io
#from PIL import Image

for face_location in face_locations:

    # Print the location of each face in this image
    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

    # You can access the actual face itself like this:
    face_image = image[top:bottom, left:right]
    #pil_image = Image.fromarray(face_image)
    #pil_image.show()
    io.imshow(face_image)
    io.show()    
```

```python
from skimage import io
import face_recognition

# Load the image file into a numpy array
image = face_recognition.load_image_file('face-image-test/unknown/u-08.jpg')
# Find all the faces in the image using the default HOG-based model.
face_locations = face_recognition.face_locations(image, model='hog')

# face_locations is now an array listing the co-ordinates of each face!
print("Found {} face(s) in this photograph.".format(len(face_locations)))

for face_location in face_locations:

    # Print the location of each face in this image
    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

    # You can access the actual face itself like this:
    face_image = image[top:bottom, left:right]
    io.imshow(face_image)
    io.show()    
```

```python
import face_recognition
from skimage import io
from matplotlib import pyplot as plt
import math

image = face_recognition.load_image_file('face-image-test/unknown/u-09.jpg')
io.imshow(image)
io.show()

face_locations = face_recognition.face_locations(image)
print("Found {} face(s) in this photograph.".format(len(face_locations)))

r_c = int(math.ceil(math.sqrt(len(face_locations))))
fig, axes = plt.subplots(nrows=r_c, ncols=r_c, figsize=(16,16))

cnt = 0
for i in(range(0, r_c)):
    for j in(range(0, r_c)):
        if(cnt >= len(face_locations)):
            top, right, bottom, left = face_locations[len(face_locations)-1]
        else:
            top, right, bottom, left = face_locations[cnt]
        
        face_image = image[top:bottom, left:right]
        
        axes[i,j].imshow(face_image) 
        axes[i,j].set_axis_off()  
        if(cnt >= len(face_locations)):
            tmp_title = "Duplicate " + str(len(face_locations))    
        else:        
            tmp_title = "Face " + str(int(cnt+1))
            
        axes[i,j].set_title(tmp_title) 
        
        cnt = cnt+1
```

## OpenCV

Haar Cascade Classifier

```python
import cv2
print(cv2.__version__)
```

```python
import sys
from matplotlib import pyplot as plt

#cascade_src = 'model/haarcascade_frontalface_default.xml'
cascade_src = 'model/haarcascade_frontalface_alt2.xml'

face_cascade = cv2.CascadeClassifier(cascade_src)

local = 'face-image-test/unknown/u-10.jpg'
img = cv2.imread(local)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)


for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)      

#cv2.imshow('object found',img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

print("number of face(s) ", len(faces))
plt.figure(figsize=(15,10))
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import sys
from matplotlib import pyplot as plt

#cascade_src = 'model/haarcascade_frontalface_default.xml'
cascade_src = 'model/haarcascade_frontalface_alt2.xml'

face_cascade = cv2.CascadeClassifier(cascade_src)

local = 'face-image-test/unknown/u-10.jpg'
img = cv2.imread(local)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=2)


for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)      

#cv2.imshow('object found',img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

print("number of face(s) ", len(faces))
plt.figure(figsize=(20,15))
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

# Extra
Car detection

```python
import cv2
import sys
from matplotlib import pyplot as plt
import numpy as np

cascade_src = 'cars.xml'
video_src = 'dataset/video1.avi'

cap = cv2.VideoCapture(video_src)
car_cascade = cv2.CascadeClassifier(cascade_src)

while True:
    
    ret, img = cap.read()
    if (type(img) == type(None)):
        exit

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cars = car_cascade.detectMultiScale(gray, 1.1, 1)

    for (x,y,w,h) in cars:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)      
        #tmp = gray[y:y+h,x:x+w]
        
    cv2.imshow('Car detection',img)

    if cv2.waitKey(33) == 27:
        break

cv2.destroyAllWindows()
```
